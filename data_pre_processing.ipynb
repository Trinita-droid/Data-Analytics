EXAMPLE 1 - GENERATE RANDOM NUMBERS 

```python
#1. IMPORT LIBRARIES :
import numpy as np #generate random numbers
import matplotlib.pyplot as plt #plot numbers
from matplotlib import style #style class is used to set the grid style
 #display the plot within jupyter notebook :
%matplotlib inline
```

```python
#2. IMPORT THE DATASET - here we have defined the dataset by generating random numbers using the random method
dataset = np.random.rand(6,2)
print(dataset)
```

```python
#3. SET THE PLOT PARAMETERS :
# set the style of the plt :
style.use('ggplot')
# labels of the coordinates : in the plot method, 'g' denotes green color.
plt.plot(dataset,'g',label='line one',linewidth = 12) #label is the legend.
# x-axis name :
plt.xlabel('Range')
#  y-axis name :
plt.ylabel('Numbers')
# title of the plot :
plt.title('First plot')
# DISPLAY THE CREATED PLOT :
plt.legend() #legend method is used to plot the graph based on set conditions 
plt.show() #display the created plot.
```

EXAMPLE 2 - A website makes analysis on how many people visit their applicaiton at a partiular time. It's a 2-D plot

```python
#1. IMPORT LIBRARIES :
import numpy as np
import pandas as pd
import matplotlib_inline
import matplotlib.pyplot as py
from matplotlib import style
```

```python
#2. IMPORT THE DATASET - here we have defined the dataset by generating random numbers using the random method
customers_using = np.random.randint(7,11,size = 5)
customers_time = np.random.randint(1,5,size = 5)
print(f"customer numbers : {customers_using} at time : {customers_time}")
```

```python
#3. SET THE PLOT PARAMETERS :
style.use('ggplot')
plt.plot(customers_time, customers_using,'pink',label = 'analysis', linewidth = 5)
# fc=plt.figure()
# fc.patch.set_facecolor('black')
plt.title('SITE TRAFFIC')
plt.xlabel('TIME')
plt.ylabel('USERS')
plt.legend()
plt.show()
```

```python
#4. CONTROLLING LINE PATTERN AND COLORS, AXIS, LABELS, LEGEND PROPERTY : can set the desired axis to interpret the result.
py.plot(customers_time, customers_using, linestyle = 'solid', color = 'black',alpha = 1)
# ALPHA IS AN ATTRIBUTE THTA CONTROLS THE TRANSPARENCY OF THE LINE, LOWER THE LINE, MORE TRANSPARENT IS THE LINE.
```

EXAMPLE 3 - GENERATE HISTOGRAM AND SCATTER PLOT FROM MATPLOTLIB

```python
#1. Import boston dataset from scikit library
from sklearn.datasets import load_boston
import matplotlib.pyplot as py
from matplotlib import style
%matplotlib inline
```

```python
#load boston dataset
boston_real_state_data = load_boston()
# xaxis
x = boston_real_state_data.target
# yaxis
y = boston_real_state_data.data
# plot histogram 
style.use('fivethirtyeight')
plt.figure(figsize=(5,5))
# plt.hist(y,bins=50)
plt.hist(y)
plt.xlabel("price")
plt.ylabel("no of houses")
plt.title("HIST")
plt.show()
# print (boston_real_state_data.DESCR)
```

DATA CLEANING USING PANDAS

```python
import pandas as pd
df1 = pd.read_excel(r"C:\Users\TJO1COB\Downloads\Customer Call List.xlsx")
df1 # before - clean it up and standardize so it all is uniform and of same dtype in a particular column, split the pincode from address.
df = df1.copy()
df
```

```python
# drop duplicates from dataset completely
df=df.drop_duplicates()
# remove columns that we do not need
# save_drop_col = df['Not_Useful_Column']
df=df.drop(columns="Not_Useful_Column")
df
```

```python
# get rid of special characters in last_name column using strip. Strip removes characters from the left and right side. but not when a character is present within a value. can specify what values need to be stripped within the brackets.
# df["Last_Name"].str.lstrip(".")
# df["Last_Name"].str.lstrip("/")
# combine all characters inside one value in a strip :  
df["Last_Name"] = df["Last_Name"].str.strip("/_...")
df
# working with phone numbers - remove the nan except alpha - numeric values
# print(df["Phone_Number"])
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"]
```

```python
# formatting/standardize the phone numbers
# df["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10] + '-')
# above implementation throws error - float object not subscriptable hence have to change each value in the column to string values.
# str(df["Phone_Number"])  - this changes whole of column into one string. , correct implementation would be :
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# df["Phone_Number"]
# df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df["Phone_Number"] 
```

```python
# split address columns and create 3 separate based off address col.
df[["Street","State","Zipcode"]] = df["Address"].str.split(',',2,expand=True)
df = df.drop(columns="Address")
df
```

```python
# pay attention to 'Y' and 'N' columns.
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','')
df["Paying Customer"]
```

```python
df
```

```python
# col 2 of Y and N
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
# df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('NaN','') - Nan cannot be removed because it is not 'NA' its just a value, hence use fillna() method
df=df.fillna('')
df
```

```python
# populate the empty cols in do_not_contact col:
df = df.replace('','N')

# # take a list of phone numbers that can be called
###PAY ATTENTION WHILE USING DROP() 
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y' :
        df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == 'Y' :
        df.drop(x,inplace=True)
df
```

```python
# reset index - drops original index and prints with a new one
df=df.reset_index(drop=True)
df
```
